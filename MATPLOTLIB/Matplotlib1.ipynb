# BASIC GRAPHS


```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]

#Title, xlabel and ylabel
plt.title('Our Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# Marking ticks (on x and y axis)
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])

#plot graph
plt.plot(x,y)
plt.show()
```

```python
x= [0,1,2,3,4]
y= [0,2,4,6,8]

#Title, xlabel and ylabel
#adding font and size
plt.title('Our Graph', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('X Axis',fontdict={'fontname':'Arial'} )
plt.ylabel('Y Axis', fontdict={'fontname':'Arial'})

# Marking ticks (on x and y axis)
plt.xticks([0,1,2,3,4,5])
plt.yticks([0,2,4,6,8,10])

#plot graph
# using label
plt.plot(x,y, label='2x', color='red', linewidth=2, marker=".", linestyle='--', markersize=10, markeredgecolor='blue')
plt.legend()
plt.show()
```

# BAR CHARTS

```python
labels= ['A','B','C']
values= [1,4,2]

bars=plt.bar(labels,values)

#hatch values on bar
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')

plt.figure(figsize=(6,4))
plt.show()
```

```python
#Other method for hue

labels = ['A', 'B', 'C']
values = [1,4,2]

plt.figure(figsize=(5,3), dpi=100)

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.savefig('barchart.png', dpi=300)

plt.show()
```
