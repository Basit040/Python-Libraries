```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
gas= pd.read_csv('gas_prices.csv')
gas
```

```python
plt.plot(gas.Year, gas.USA)
plt.show()
```

```python
#Set figure size
plt.figure(figsize=(8,5))

#Set Title
plt.title("Gas Prices over time (in US Dollar)")

#Plot for different countries
plt.plot(gas.Year, gas.USA, 'b.-', label="USA")
plt.plot(gas.Year, gas.Canada, 'r.-', label="Canada")
plt.plot(gas.Year, gas['South Korea'], 'g.-', label="South Korea")
plt.plot(gas.Year, gas.Australia, 'y.-', label="Australia")

#Set ticks
# ::3 means start from initial value and skip 3 values and go to the end
plt.xticks(gas.Year[::3])

# Add labels
plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.legend()
plt.show()
```

```python
# We will count players with left foot
fifa.loc[fifa['Preferred Foot']=="Left"].count()
```

```python
# we want only one number so
fifa.loc[fifa['Preferred Foot']=="Left"].count()[0]
```

```python
# We will create two variables for left and right foot
left= fifa.loc[fifa['Preferred Foot']=="Left"].count()[0]
right=fifa.loc[fifa['Preferred Foot']=="Right"].count()[0]

# Set labels and color
labels= ['Left','Right']
colors= ['#abcdef','#aabbcc']

#pie chart with autopct (represent % )
plt.pie([left,right], labels=labels,colors=colors,autopct='%.2f%%')

#Set title
plt.title('Foot Preferences of Fifa Players')

plt.show()
```

```python
fifa.Weight
```

```python
# we are creating weight variable in which "lbs will be deleted" and also null value will be ignored
# x.strip('lbs') will delete lbs value
# int(x.strip('lbs')) convert in integar
# if type(x)==str "will ignore non numeric value"

fifa.Weight= [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight[0]
```

```python
plt.figure(figsize=(8,5), dpi=100)

#Use ggplot style
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

# Make different categories of weights
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa[fifa.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']

#we want to explode under and over weight, so values are 0.4
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')
plt.show()
```

```python
# How many players play for barcelona
barcelona= fifa.loc[fifa.Club=='FC Barcelona']['Overall']
barcelona
```

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#bp = plt.boxplot([barcelona, madrid, revs], labels=['a','b','c'], boxprops=dict(facecolor='red'))
bp = plt.boxplot([barcelona, madrid, revs], labels=['FC Barcelona','Real Madrid','NE Revolution'], patch_artist=True, medianprops={'linewidth': 2})

plt.title('Professional Soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')

# Loop to add color etc on all boxes easily
for box in bp['boxes']:
    # change outline color
    box.set(color='#4286f4', linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
    # change hatch
    #box.set(hatch = '/')
    
plt.show()
```
